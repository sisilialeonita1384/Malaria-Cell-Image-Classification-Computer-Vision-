```python
# ============================================================
# EfficientNetV2 - otsu
# ============================================================

import os
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
from torchvision.models import efficientnet_v2_s, EfficientNet_V2_S_Weights

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, roc_curve, auc
)

import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import numpy as np

# ============================================================
# TIMER GLOBAL
# ============================================================
cell_start_time = time.time()

# ============================================================
# KONFIGURASI
# ============================================================
DATA_ROOT = "/kaggle/input/processed-images/processed_images"
PREPROCESS_MODE = "otsu"  
BATCH_SIZE = 64
LEARNING_RATE = 5e-5
WEIGHT_DECAY = 1e-4
PATIENCE = 5
MAX_EPOCHS = 30

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ============================================================
# DATA AUGMENTATION
# ============================================================
train_transform = transforms.Compose([
    transforms.Grayscale(3),
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.15, contrast=0.15),
    transforms.ToTensor(),
    transforms.RandomErasing(p=0.25, scale=(0.02, 0.15))
])

val_transform = transforms.Compose([
    transforms.Grayscale(3),
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

def filter_preprocess(path):
    return PREPROCESS_MODE in path

# ============================================================
# LOAD DATASET
# ============================================================
full_dataset = datasets.ImageFolder(
    root=DATA_ROOT,
    transform=None,
    is_valid_file=filter_preprocess
)

class_names = full_dataset.classes
print("Classes:", class_names)

total_len = len(full_dataset)
train_len = int(0.7 * total_len)
val_len = int(0.15 * total_len)
test_len = total_len - train_len - val_len

train_indices, val_indices, test_indices = torch.utils.data.random_split(
    range(total_len), [train_len, val_len, test_len]
)

# Custom dataset
class CustomSubset(torch.utils.data.Dataset):
    def __init__(self, dataset, indices, transform):
        self.dataset = dataset
        self.indices = indices
        self.transform = transform
    
    def __len__(self):
        return len(self.indices)
    
    def __getitem__(self, idx):
        img_path, label = self.dataset.samples[self.indices[idx]]
        img = self.dataset.loader(img_path)
        if self.transform:
            img = self.transform(img)
        return img, label

train_set = CustomSubset(full_dataset, train_indices.indices, train_transform)
val_set = CustomSubset(full_dataset, val_indices.indices, val_transform)
test_set = CustomSubset(full_dataset, test_indices.indices, val_transform)

# DataLoader optimized
train_loader = DataLoader(
    train_set, 
    batch_size=BATCH_SIZE, 
    shuffle=True, 
    num_workers=2,
    pin_memory=True,
    persistent_workers=True
)

val_loader = DataLoader(
    val_set, 
    batch_size=BATCH_SIZE * 2, 
    shuffle=False, 
    num_workers=2,
    pin_memory=True,
    persistent_workers=True
)

test_loader = DataLoader(
    test_set, 
    batch_size=BATCH_SIZE * 2, 
    shuffle=False, 
    num_workers=2,
    pin_memory=True,
    persistent_workers=True
)

print(f"Train: {len(train_set)}, Val: {len(val_set)}, Test: {len(test_set)}")
print(f"Batches - Train: {len(train_loader)}, Val: {len(val_loader)}, Test: {len(test_loader)}")

# ============================================================
# MODEL DENGAN DROPOUT
# ============================================================
print("\nLoading EfficientNetV2-S...")
weights = EfficientNet_V2_S_Weights.IMAGENET1K_V1
model = efficientnet_v2_s(weights=weights)

# Modifikasi classifier dengan dropout
num_features = model.classifier[1].in_features
model.classifier = nn.Sequential(
    nn.Dropout(p=0.3, inplace=True),
    nn.Linear(num_features, 512),
    nn.ReLU(inplace=True),
    nn.Dropout(p=0.4),
    nn.Linear(512, len(class_names))
)

model = model.to(device)
print("Model loaded successfully!")

# ============================================================
# LOSS & OPTIMIZER
# ============================================================
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

# Learning rate scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='max', factor=0.5, patience=3
)

# ============================================================
# TRAINING FUNCTION
# ============================================================
def run_epoch(model, loader, optimizer=None):
    train_mode = optimizer is not None
    model.train() if train_mode else model.eval()

    running_loss = 0.0
    correct = 0
    total = 0
    
    pbar = tqdm(loader, desc="Train" if train_mode else "Val", leave=True)

    with torch.set_grad_enabled(train_mode):
        for x, y in pbar:
            x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
            
            if train_mode:
                optimizer.zero_grad(set_to_none=True)
            
            out = model(x)
            loss = criterion(out, y)
            
            if train_mode:
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
            
            running_loss += loss.item() * x.size(0)
            preds = torch.argmax(out, dim=1)
            correct += (preds == y).sum().item()
            total += y.size(0)

            pbar.set_postfix({
                "loss": f"{running_loss/total:.4f}",
                "acc": f"{correct/total:.4f}"
            })
    
    return running_loss / total, correct / total

# ============================================================
# TRAINING LOOP
# ============================================================
best_val_acc = 0.0
patience_counter = 0
model_file = f"best_effnetv2s_{PREPROCESS_MODE}_antioverfit.pth"

train_loss_history, val_loss_history = [], []
train_acc_history, val_acc_history = [], []

print("\n" + "="*60)
print("STARTING TRAINING")
print("="*60)
train_start = time.time()

for epoch in range(1, MAX_EPOCHS + 1):
    epoch_start = time.time()
    
    print(f"\nEpoch {epoch}/{MAX_EPOCHS}")
    print("-" * 40)
    
    train_loss, train_acc = run_epoch(model, train_loader, optimizer)
    val_loss, val_acc = run_epoch(model, val_loader)
    
    scheduler.step(val_acc)
    current_lr = optimizer.param_groups[0]['lr']
    
    train_loss_history.append(train_loss)
    val_loss_history.append(val_loss)
    train_acc_history.append(train_acc)
    val_acc_history.append(val_acc)
    
    epoch_time = time.time() - epoch_start
    
    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.4f}")
    print(f"Gap: {(train_acc - val_acc):.4f} | LR: {current_lr:.2e} | Time: {epoch_time:.1f}s")
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), model_file)
        patience_counter = 0
        print(f"✓ Saved! Best Val Acc: {best_val_acc:.4f}")
    else:
        patience_counter += 1
        print(f"✗ No improvement ({patience_counter}/{PATIENCE})")
        
        if patience_counter >= PATIENCE:
            print("\n EARLY STOPPING")
            break

train_end = time.time()
training_time = train_end - train_start
print(f"\n{'='*60}")
print(f"Training Done: {training_time/60:.1f} min ({training_time:.1f}s)")
print(f"Best Val Acc: {best_val_acc:.4f}")
print(f"{'='*60}")

# Clear cache
torch.cuda.empty_cache()

# ============================================================
# TESTING
# ============================================================
print("\n" + "="*60)
print("TESTING")
print("="*60)

model.load_state_dict(torch.load(model_file, map_location=device))
model.eval()

all_labels, all_preds, all_probs = [], [], []
test_start = time.time()

with torch.no_grad():
    for x, y in tqdm(test_loader, desc="Testing"):
        x = x.to(device, non_blocking=True)
        out = model(x)
        probs = torch.softmax(out, dim=1).cpu().numpy()
        preds = np.argmax(probs, axis=1)
        
        all_labels.extend(y.numpy())
        all_preds.extend(preds)
        
        if probs.shape[1] == 2:
            all_probs.extend(probs[:, 1])
        else:
            all_probs.extend(probs.tolist())

test_end = time.time()
testing_time = test_end - test_start
total_runtime = time.time() - cell_start_time

# ============================================================
# METRICS
# ============================================================
acc = accuracy_score(all_labels, all_preds)
prec = precision_score(all_labels, all_preds, zero_division=0)
rec = recall_score(all_labels, all_preds, zero_division=0)
f1 = f1_score(all_labels, all_preds, zero_division=0)

if isinstance(all_probs[0], (float, np.floating)):
    auc_score = roc_auc_score(all_labels, all_probs)
else:
    auc_score = None

print("\n" + "="*60)
print("TEST RESULTS")
print("="*60)
print(f"Preprocessing  : {PREPROCESS_MODE}")
print(f"Accuracy       : {acc:.4f}")
print(f"Precision      : {prec:.4f}")
print(f"Recall         : {rec:.4f}")
print(f"F1-Score       : {f1:.4f}")
print(f"AUC / ROC      : {auc_score:.4f}" if auc_score else "AUC / ROC      : N/A")
print(f"Training Time  : {training_time/60:.1f} min")
print(f"Testing Time   : {testing_time:.1f} sec")
print(f"Total Runtime  : {total_runtime/60:.1f} min")
print("="*60)

# ============================================================
# VISUALIZATIONS
# ============================================================

# 1. Confusion Matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(all_labels, all_preds)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", 
            xticklabels=class_names, yticklabels=class_names, cbar_kws={'label': 'Count'})
plt.title(f"Confusion Matrix - {PREPROCESS_MODE}", fontsize=14, fontweight='bold')
plt.xlabel("Predicted", fontsize=12)
plt.ylabel("True", fontsize=12)
plt.tight_layout()
plt.show()

# 2. ROC Curve
if auc_score is not None:
    fpr, tpr, _ = roc_curve(all_labels, all_probs)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, lw=2, color='#2E86AB', label=f'AUC = {auc_score:.4f}')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', alpha=0.5, label='Random')
    plt.xlabel("False Positive Rate", fontsize=12)
    plt.ylabel("True Positive Rate", fontsize=12)
    plt.title(f"ROC Curve - {PREPROCESS_MODE}", fontsize=14, fontweight='bold')
    plt.legend(fontsize=11)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# 3. Learning Curves
epochs = range(1, len(train_loss_history) + 1)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Loss curve
axes[0].plot(epochs, train_loss_history, 'o-', color='#2E86AB', label="Train Loss", markersize=5, linewidth=2)
axes[0].plot(epochs, val_loss_history, 's-', color='#A23B72', label="Val Loss", markersize=5, linewidth=2)
axes[0].set_xlabel("Epoch", fontsize=12)
axes[0].set_ylabel("Loss", fontsize=12)
axes[0].set_title("Loss per Epoch", fontsize=13, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].grid(alpha=0.3)

# Accuracy curve
axes[1].plot(epochs, train_acc_history, 'o-', color='#2E86AB', label="Train Acc", markersize=5, linewidth=2)
axes[1].plot(epochs, val_acc_history, 's-', color='#A23B72', label="Val Acc", markersize=5, linewidth=2)
axes[1].set_xlabel("Epoch", fontsize=12)
axes[1].set_ylabel("Accuracy", fontsize=12)
axes[1].set_title("Accuracy per Epoch", fontsize=13, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 4. Overfitting Analysis
print("\n" + "="*60)
print("OVERFITTING ANALYSIS")
print("="*60)
final_gap = train_acc_history[-1] - val_acc_history[-1]
print(f"Final Train Acc: {train_acc_history[-1]:.4f}")
print(f"Final Val Acc  : {val_acc_history[-1]:.4f}")
print(f"Accuracy Gap   : {final_gap:.4f}")

if final_gap < 0.05:
    print("Status: MODEL GENERALISASI DENGAN BAIK")
elif final_gap < 0.10:
    print("Status: SEDIKIT OVERFITTING (masih acceptable)")
else:
    print("Status: OVERFITTING DETECTED")
print("="*60)
```

```python
# ============================================================
# EfficientNetV2 - otsu_canny
# ============================================================

import os
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
from torchvision.models import efficientnet_v2_s, EfficientNet_V2_S_Weights

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, roc_curve, auc
)

import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import numpy as np

# ============================================================
# TIMER GLOBAL
# ============================================================
cell_start_time = time.time()

# ============================================================
# KONFIGURASI
# ============================================================
DATA_ROOT = "/kaggle/input/processed-images/processed_images"
PREPROCESS_MODE = "otsu_canny"  
BATCH_SIZE = 64
LEARNING_RATE = 5e-5
WEIGHT_DECAY = 1e-4
PATIENCE = 5
MAX_EPOCHS = 30

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ============================================================
# DATA AUGMENTATION
# ============================================================
train_transform = transforms.Compose([
    transforms.Grayscale(3),
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.15, contrast=0.15),
    transforms.ToTensor(),
    transforms.RandomErasing(p=0.25, scale=(0.02, 0.15))
])

val_transform = transforms.Compose([
    transforms.Grayscale(3),
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

def filter_preprocess(path):
    return PREPROCESS_MODE in path

# ============================================================
# LOAD DATASET
# ============================================================
full_dataset = datasets.ImageFolder(
    root=DATA_ROOT,
    transform=None,
    is_valid_file=filter_preprocess
)

class_names = full_dataset.classes
print("Classes:", class_names)

total_len = len(full_dataset)
train_len = int(0.7 * total_len)
val_len = int(0.15 * total_len)
test_len = total_len - train_len - val_len

train_indices, val_indices, test_indices = torch.utils.data.random_split(
    range(total_len), [train_len, val_len, test_len]
)

# Custom dataset
class CustomSubset(torch.utils.data.Dataset):
    def __init__(self, dataset, indices, transform):
        self.dataset = dataset
        self.indices = indices
        self.transform = transform
    
    def __len__(self):
        return len(self.indices)
    
    def __getitem__(self, idx):
        img_path, label = self.dataset.samples[self.indices[idx]]
        img = self.dataset.loader(img_path)
        if self.transform:
            img = self.transform(img)
        return img, label

train_set = CustomSubset(full_dataset, train_indices.indices, train_transform)
val_set = CustomSubset(full_dataset, val_indices.indices, val_transform)
test_set = CustomSubset(full_dataset, test_indices.indices, val_transform)

# DataLoader optimized
train_loader = DataLoader(
    train_set, 
    batch_size=BATCH_SIZE, 
    shuffle=True, 
    num_workers=2,
    pin_memory=True,
    persistent_workers=True
)

val_loader = DataLoader(
    val_set, 
    batch_size=BATCH_SIZE * 2, 
    shuffle=False, 
    num_workers=2,
    pin_memory=True,
    persistent_workers=True
)

test_loader = DataLoader(
    test_set, 
    batch_size=BATCH_SIZE * 2, 
    shuffle=False, 
    num_workers=2,
    pin_memory=True,
    persistent_workers=True
)

print(f"Train: {len(train_set)}, Val: {len(val_set)}, Test: {len(test_set)}")
print(f"Batches - Train: {len(train_loader)}, Val: {len(val_loader)}, Test: {len(test_loader)}")

# ============================================================
# MODEL DENGAN DROPOUT
# ============================================================
print("\nLoading EfficientNetV2-S...")
weights = EfficientNet_V2_S_Weights.IMAGENET1K_V1
model = efficientnet_v2_s(weights=weights)

# Modifikasi classifier dengan dropout
num_features = model.classifier[1].in_features
model.classifier = nn.Sequential(
    nn.Dropout(p=0.3, inplace=True),
    nn.Linear(num_features, 512),
    nn.ReLU(inplace=True),
    nn.Dropout(p=0.4),
    nn.Linear(512, len(class_names))
)

model = model.to(device)
print("Model loaded successfully!")

# ============================================================
# LOSS & OPTIMIZER
# ============================================================
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

# Learning rate scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='max', factor=0.5, patience=3
)

# ============================================================
# TRAINING FUNCTION
# ============================================================
def run_epoch(model, loader, optimizer=None):
    train_mode = optimizer is not None
    model.train() if train_mode else model.eval()

    running_loss = 0.0
    correct = 0
    total = 0
    
    pbar = tqdm(loader, desc="Train" if train_mode else "Val", leave=True)

    with torch.set_grad_enabled(train_mode):
        for x, y in pbar:
            x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
            
            if train_mode:
                optimizer.zero_grad(set_to_none=True)
            
            out = model(x)
            loss = criterion(out, y)
            
            if train_mode:
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
            
            running_loss += loss.item() * x.size(0)
            preds = torch.argmax(out, dim=1)
            correct += (preds == y).sum().item()
            total += y.size(0)

            pbar.set_postfix({
                "loss": f"{running_loss/total:.4f}",
                "acc": f"{correct/total:.4f}"
            })
    
    return running_loss / total, correct / total

# ============================================================
# TRAINING LOOP
# ============================================================
best_val_acc = 0.0
patience_counter = 0
model_file = f"best_effnetv2s_{PREPROCESS_MODE}_antioverfit.pth"

train_loss_history, val_loss_history = [], []
train_acc_history, val_acc_history = [], []

print("\n" + "="*60)
print("STARTING TRAINING")
print("="*60)
train_start = time.time()

for epoch in range(1, MAX_EPOCHS + 1):
    epoch_start = time.time()
    
    print(f"\nEpoch {epoch}/{MAX_EPOCHS}")
    print("-" * 40)
    
    train_loss, train_acc = run_epoch(model, train_loader, optimizer)
    val_loss, val_acc = run_epoch(model, val_loader)
    
    scheduler.step(val_acc)
    current_lr = optimizer.param_groups[0]['lr']
    
    train_loss_history.append(train_loss)
    val_loss_history.append(val_loss)
    train_acc_history.append(train_acc)
    val_acc_history.append(val_acc)
    
    epoch_time = time.time() - epoch_start
    
    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.4f}")
    print(f"Gap: {(train_acc - val_acc):.4f} | LR: {current_lr:.2e} | Time: {epoch_time:.1f}s")
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), model_file)
        patience_counter = 0
        print(f"✓ Saved! Best Val Acc: {best_val_acc:.4f}")
    else:
        patience_counter += 1
        print(f"✗ No improvement ({patience_counter}/{PATIENCE})")
        
        if patience_counter >= PATIENCE:
            print("\n EARLY STOPPING")
            break

train_end = time.time()
training_time = train_end - train_start
print(f"\n{'='*60}")
print(f"Training Done: {training_time/60:.1f} min ({training_time:.1f}s)")
print(f"Best Val Acc: {best_val_acc:.4f}")
print(f"{'='*60}")

# Clear cache
torch.cuda.empty_cache()

# ============================================================
# TESTING
# ============================================================
print("\n" + "="*60)
print("TESTING")
print("="*60)

model.load_state_dict(torch.load(model_file, map_location=device))
model.eval()

all_labels, all_preds, all_probs = [], [], []
test_start = time.time()

with torch.no_grad():
    for x, y in tqdm(test_loader, desc="Testing"):
        x = x.to(device, non_blocking=True)
        out = model(x)
        probs = torch.softmax(out, dim=1).cpu().numpy()
        preds = np.argmax(probs, axis=1)
        
        all_labels.extend(y.numpy())
        all_preds.extend(preds)
        
        if probs.shape[1] == 2:
            all_probs.extend(probs[:, 1])
        else:
            all_probs.extend(probs.tolist())

test_end = time.time()
testing_time = test_end - test_start
total_runtime = time.time() - cell_start_time

# ============================================================
# METRICS
# ============================================================
acc = accuracy_score(all_labels, all_preds)
prec = precision_score(all_labels, all_preds, zero_division=0)
rec = recall_score(all_labels, all_preds, zero_division=0)
f1 = f1_score(all_labels, all_preds, zero_division=0)

if isinstance(all_probs[0], (float, np.floating)):
    auc_score = roc_auc_score(all_labels, all_probs)
else:
    auc_score = None

print("\n" + "="*60)
print("TEST RESULTS")
print("="*60)
print(f"Preprocessing  : {PREPROCESS_MODE}")
print(f"Accuracy       : {acc:.4f}")
print(f"Precision      : {prec:.4f}")
print(f"Recall         : {rec:.4f}")
print(f"F1-Score       : {f1:.4f}")
print(f"AUC / ROC      : {auc_score:.4f}" if auc_score else "AUC / ROC      : N/A")
print(f"Training Time  : {training_time/60:.1f} min")
print(f"Testing Time   : {testing_time:.1f} sec")
print(f"Total Runtime  : {total_runtime/60:.1f} min")
print("="*60)

# ============================================================
# VISUALIZATIONS
# ============================================================

# 1. Confusion Matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(all_labels, all_preds)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", 
            xticklabels=class_names, yticklabels=class_names, cbar_kws={'label': 'Count'})
plt.title(f"Confusion Matrix - {PREPROCESS_MODE}", fontsize=14, fontweight='bold')
plt.xlabel("Predicted", fontsize=12)
plt.ylabel("True", fontsize=12)
plt.tight_layout()
plt.show()

# 2. ROC Curve
if auc_score is not None:
    fpr, tpr, _ = roc_curve(all_labels, all_probs)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, lw=2, color='#2E86AB', label=f'AUC = {auc_score:.4f}')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', alpha=0.5, label='Random')
    plt.xlabel("False Positive Rate", fontsize=12)
    plt.ylabel("True Positive Rate", fontsize=12)
    plt.title(f"ROC Curve - {PREPROCESS_MODE}", fontsize=14, fontweight='bold')
    plt.legend(fontsize=11)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# 3. Learning Curves
epochs = range(1, len(train_loss_history) + 1)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Loss curve
axes[0].plot(epochs, train_loss_history, 'o-', color='#2E86AB', label="Train Loss", markersize=5, linewidth=2)
axes[0].plot(epochs, val_loss_history, 's-', color='#A23B72', label="Val Loss", markersize=5, linewidth=2)
axes[0].set_xlabel("Epoch", fontsize=12)
axes[0].set_ylabel("Loss", fontsize=12)
axes[0].set_title("Loss per Epoch", fontsize=13, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].grid(alpha=0.3)

# Accuracy curve
axes[1].plot(epochs, train_acc_history, 'o-', color='#2E86AB', label="Train Acc", markersize=5, linewidth=2)
axes[1].plot(epochs, val_acc_history, 's-', color='#A23B72', label="Val Acc", markersize=5, linewidth=2)
axes[1].set_xlabel("Epoch", fontsize=12)
axes[1].set_ylabel("Accuracy", fontsize=12)
axes[1].set_title("Accuracy per Epoch", fontsize=13, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 4. Overfitting Analysis
print("\n" + "="*60)
print("OVERFITTING ANALYSIS")
print("="*60)
final_gap = train_acc_history[-1] - val_acc_history[-1]
print(f"Final Train Acc: {train_acc_history[-1]:.4f}")
print(f"Final Val Acc  : {val_acc_history[-1]:.4f}")
print(f"Accuracy Gap   : {final_gap:.4f}")

if final_gap < 0.05:
    print("Status: MODEL GENERALISASI DENGAN BAIK")
elif final_gap < 0.10:
    print("Status: SEDIKIT OVERFITTING (masih acceptable)")
else:
    print("Status: OVERFITTING DETECTED")
print("="*60)
```
